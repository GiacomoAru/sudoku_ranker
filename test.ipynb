```python
import json
from pathlib import Path
from typing import Any, Callable


def modifica_json_cartella(
    cartella: str | Path,
    modifica: Callable[[Any], Any],
    *,
    ricorsivo: bool = False,
    indent: int = 4,
) -> None:
    """
    Apre tutti i file JSON nella cartella, applica `modifica`
    al contenuto e sovrascrive il file originale.

    `modifica` deve ricevere il contenuto JSON e restituire
    il nuovo contenuto da salvare.
    """
    cartella = Path(cartella)

    if not cartella.is_dir():
        raise NotADirectoryError(f"Cartella non valida: {cartella}")

    files = cartella.rglob("*.json") if ricorsivo else cartella.glob("*.json")

    for file_path in files:
        try:
            with file_path.open("r", encoding="utf-8") as file:
                dati = json.load(file)

            nuovi_dati = modifica(dati)

            with file_path.open("w", encoding="utf-8") as file:
                json.dump(
                    nuovi_dati,
                    file,
                    indent=indent,
                    ensure_ascii=False,
                )

            print(f"Modificato: {file_path}")

        except (json.JSONDecodeError, OSError, TypeError) as errore:
            print(f"Errore in {file_path}: {errore}")
```

```python
def aggiungi_versione(dati):
    dati['metadata']['tag'] = 'campain'

    return dati


modifica_json_cartella(
    "./sudoku_data/puzzles",
    aggiungi_versione,
)
```
